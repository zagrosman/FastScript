```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load your CSV data into a DataFrame
df = pd.read_csv('data.csv')

# Create a scatter plot using Seaborn
plt.figure(figsize=(10, 6), dpi=700)  # Set a higher DPI for high resolution
sns.scatterplot(x=np.arange(len(df)), y='Score', data=df, hue='Score', s=80, alpha=0.7, palette='viridis')

# Set x-axis labels and major ticks
plt.xticks(np.arange(0, len(df), 100), np.arange(0, len(df), 100))

# Set plot labels and title
plt.xlabel('Total number of ligands (686)')
plt.ylabel('Score (kcal.mol)')
plt.title('Molecular Docking Binding Energies')

# Save the plot in high resolution (adjust the file format if needed)
plt.savefig('scatter_plot_high_res.png', bbox_inches='tight')

# Show the plot
plt.show()
```
